```python
import pandas as pd
import numpy as np
df = pd.read_csv('data/vginsights_gameloft_data.csv')
df.head()
```

```python
df["Active Players (now)"] = df["Active Players (now)"].str.replace(",","").astype(float)
df["Active Players (24h peak)"] = df["Active Players (24h peak)"].str.replace(",","").astype(float) 
df["Units Sold"] = df["Units Sold"].apply(lambda x: 
                                        float(x.replace("m","")) * 1000000 if "m" in x else 
                                        float(x.replace("k","")) * 1000 if "k" in x else 
                                        float(x.replace(",","")) if "," in x else
                                        float(x))

```

```python
df.head()
```

```python
df.to_csv('data/vginsights_gameloft_data_cleaned.csv', index=False)
```
